```python
import folium
from pyproj import Transformer
from folium import DivIcon

# Umrechner von LV95 (EPSG:2056) nach WGS84 (EPSG:4326)
transformer = Transformer.from_crs("EPSG:2056", "EPSG:4326", always_xy=True)

# Umwandlung für alle Zeilen in df_karte
df_karte_tram["Longitude"], df_karte_tram["Latitude"] = zip(*df_karte_tram.apply(lambda row: transformer.transform(row["E"], row["N"]), axis=1))

# Mittelpunkt der Karte berechnen
map_center = [df_karte_tram["Latitude"].mean(), df_karte_tram["Longitude"].mean()]
karte = folium.Map(location=map_center, zoom_start=13)

# Stationen auf der Karte als Marker hinzufügen
for _, row in df_karte_tram.iterrows():
        popup_text = f"""
    <b>Station:</b> {row.get("Name", "Unbekannt")}<br>
    <b>Höhe:</b> {row.get("H", "Keine Höhenangabe")}<br>
    <b>Verspätung:</b> {row.get("Verspätung", "Keine Daten")}
    """
        folium.Marker(
            location=[row["Latitude"], row["Longitude"]],
            popup=folium.Popup(popup_text, max_width=400),
            icon=folium.Icon(icon="info-sign")
        ).add_to(karte)

        # Wichtige Station hervorheben (Rot)
        folium.Marker(
            location=[47.3776, 8.5304],
            popup="Messstation Kaserne",
            icon=folium.Icon(icon="cloud", color="red")  # **Hervorheben in Rot**
        ).add_to(karte)

        folium.Circle(
            location=[47.3776, 8.5304],
            radius=5500,  # Radius in Metern
            color="black",  # Randfarbe
            fill=True,
            fill_color="grey",  # Füllfarbe
            fill_opacity=0.0025  # Transparenz
        ).add_to(karte)

        folium.Marker(
            location=[47.3776, 8.5304],
            icon=DivIcon(
            icon_size=(150, 36),
            icon_anchor=(150, 300),
            html='<div style="font-size: 14pt; color: black;">5.5 km Radius</div>',
    )
).add_to(karte)

# Karte als HTML speichern
karte.save("zuerich_stationen.html")

# Falls du in Jupyter Notebook arbeitest, um die Karte direkt anzuzeigen:
karte
```

```python
import folium
from pyproj import Transformer

# Umrechner von LV95 (EPSG:2056) nach WGS84 (EPSG:4326)
transformer = Transformer.from_crs("EPSG:2056", "EPSG:4326", always_xy=True)

# 🟦 Umwandlung für Tramstationen
df_karte_tram["Longitude"], df_karte_tram["Latitude"] = zip(*df_karte_tram.apply(lambda row: transformer.transform(row["E"], row["N"]), axis=1))

# 🟥 Umwandlung für Busstationen
df_karte_bus["Longitude"], df_karte_bus["Latitude"] = zip(*df_karte_bus.apply(lambda row: transformer.transform(row["E"], row["N"]), axis=1))

# 🛑 Ermitteln von einzigartigen und gemeinsamen Stationen
tram_coords = set(zip(df_karte_tram["Latitude"], df_karte_tram["Longitude"]))
bus_coords = set(zip(df_karte_bus["Latitude"], df_karte_bus["Longitude"]))

# 🎯 Gemeinsame Stationen (Grün)
gemeinsame_coords = tram_coords & bus_coords
# 🟦 Nur Tram-Stationen (Blau)
nur_tram_coords = tram_coords - bus_coords
# 🟥 Nur Bus-Stationen (Rot)
nur_bus_coords = bus_coords - tram_coords

# Mittelpunkt der Karte berechnen
map_center = [df_karte_tram["Latitude"].mean(), df_karte_tram["Longitude"].mean()]
karte = folium.Map(location=map_center, zoom_start=13)

# 🟦 Marker für Tram-Stationen (Blau)
for _, row in df_karte_tram.iterrows():
    coord = (row["Latitude"], row["Longitude"])
    if coord in gemeinsame_coords:
        continue  # Wird später als grüner Marker gesetzt
    icon_color = "blue" if coord in nur_tram_coords else "green"

    popup_text = f"""
    <b>Station:</b> {row.get("Name", "Unbekannt")}<br>
    <b>Höhe:</b> {row.get("H", "Keine Höhenangabe")}<br>
    <b>Verspätung:</b> {row.get("Verspätung", "Keine Daten")}
    """
    folium.Marker(
        location=[row["Latitude"], row["Longitude"]],
        popup=folium.Popup(popup_text, max_width=400),
        icon=folium.Icon(icon="info-sign", color=icon_color)
    ).add_to(karte)

# 🟥 Marker für Bus-Stationen (Rot)
for _, row in df_karte_bus.iterrows():
    coord = (row["Latitude"], row["Longitude"])
    if coord in gemeinsame_coords:
        continue  # Wird später als grüner Marker gesetzt
    icon_color = "red" if coord in nur_bus_coords else "green"

    popup_text = f"""
    <b>Station:</b> {row.get("Name", "Unbekannt")}<br>
    <b>Höhe:</b> {row.get("H", "Keine Höhenangabe")}<br>
    <b>Verspätung:</b> {row.get("Verspätung", "Keine Daten")}
    """
    folium.Marker(
        location=[row["Latitude"], row["Longitude"]],
        popup=folium.Popup(popup_text, max_width=400),
        icon=folium.Icon(icon="info-sign", color=icon_color)
    ).add_to(karte)

# ✅ Gemeinsame Stationen in Grün markieren
for lat, lon in gemeinsame_coords:
    folium.Marker(
        location=[lat, lon],
        popup="Gemeinsame Tram- & Busstation",
        icon=folium.Icon(icon="info-sign", color="green")
    ).add_to(karte)

# 🔴 Messstation Kaserne als Extra-Marker
folium.Marker(
    location=[47.3776, 8.5304],
    popup="Messstation Kaserne",
    icon=folium.Icon(icon="cloud", color="red")
).add_to(karte)

# Karte speichern
karte.save("zuerich_stationen.html")

# Falls du in Jupyter Notebook arbeitest:
karte
```
